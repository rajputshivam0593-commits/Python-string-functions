String Functions
1. Capitalize()
   It  convert first letter of string to uppercase. 

```python
a="shivam"
print(a.capitalize())
```

2. upper()
   It  convert all lowercase letter of string to uppercase.

```python
a="shivam"
print(a.upper())
```

3. lower()
  It convert all uppercase letter of string to lowercase.

```python
a="SHIVAM"
print(a.lower())
```

4. title()
 It covert the first character of each word is uppercase.

```python
a="my name is shivam"
print(a.title())
```

5. swapcase()
  It convert uppercase to lowercase and lowercase to uppercase.

```python
a="SHIVAM"
b="shivam"
print(a.swapcase())
print(b.swapcase())
```

6. count()
   It counts repeat words.

```python
text="I am Shivam Rajput"
print(text.count("a"))
```

7. find()
   It finds the position of word or letter.

```python
text="I am Shivam Rajput"
print(text.find("S"))
```

8. startswith()
   It checks whether a string begin with specific sequence of characters and returns a simple True or False.

```python
text="I am Shivam Rajput"
print(text.startswith("I"))
```

9. endswith()
     It checks whether a string end with specific sequence of characters and returns a simple True or False.

```python
text="I am Shivam Rajput"
print(text.endswith("Rajput"))
```

10. strip()
    Its primary job is to remove leading and trailing whitespaces(spaces,tabs and newlines) from a string.

```python
text=" python_master_99 \n"
print(text.strip())
```

11. replace()
    It searches for a specificed substring and replaces it with a new string.

```python
text="I am Shivam Rajput"
print(text.replace("Shivam","xyz"))
```

12. join()
    It allows to merge a list of strings into a single string.

```python
words=["python","is","fun"]
print(" ".join(words))
```

13. split()
    It takes single string and breaks it apart into a list of smaller strings.

```python
data="apple,banana,orange"
fruits=data.split(",")
print(fruits)
```

14. isalpha()
    It’s your go-to tool when you need to validate that a user hasn't accidentally typed numbers or symbols into a name field.
   

```python
name = "Gemini"
print(name.isalpha())
```

15. isnumeric()
    It includes standard digits (0-9), but it also covers characters from other languages and special mathematical symbols like exponents or Roman numerals.

```python
age = "25"
print(age.isnumeric())
```

16. isspace()
    It is particularly useful when you need to clean up data or check if a user has submitted a "blank" input that only contains spaces or tabs.

```python
mixed = "\t\n \r"
print(mixed.isspace())
```

17. center()
    Gemini said
The center() method is used to center-align a string by padding it with a specified character (which defaults to a standard space) on both sides.

```python
header = "CHAPTER 1"
print(header.center(30, "-"))
```

18. ljust()
    It pushes the original text to the start (left) of the string and fills the remaining space to the right with a padding character.

```python
item = "Total"
print(item.ljust(15, "."))
```

19. rjust()
    It returns a new string of a specified length where the original string is pushed to the right, and the left side is padded with a specific character (which defaults to a space).

```python
a="hi"
print(a.rjust(7,"*"))
```

20. zfill()
    It adds zeros (0) to the beginning of a string until it reaches a specified length.
While it is commonly used for numbers, it is a string method, so it works on any text.

```python
number = "42"
print(number.zfill(5))
```

21. rfind()
    It is used to search for the last occurrence of a specified substring within a string.

```python
text = "The cat sat on the mat with another cat."
print(text.rfind("cat"))
```

22. isascii()
    ASCII (American Standard Code for Information Interchange) includes characters with Unicode code points in the range U+0000 to U+007F. This covers standard English letters (A-Z, a-z), numbers (0-9), common punctuation, and control characters like newlines.

```python
text = "Hello 123!"
print(text.isascii())
```

23. isdecimal()
    It specifically looks for characters that can be used to form numbers in base 10 (the digits 0-9).

```python
number = "12345"
print(number.isdecimal())
```

24. isidentifier()
    it tells you if a string can be used as a variable name, function name, or class name.

```python
print("my_variable".isidentifier())
print("123user".isidentifier())
```

25. expandtabs()
    The method calculates how many spaces are needed to reach the next "tab stop." It doesn't just swap every \t for 4 spaces; it ensures the text following the tab starts at a multiple of the tab size.

```python
text = "Name\tLanguage"
print(text.expandtabs())
```

26. index()
    The method returns an integer representing the index of the substring. Like all Python strings, the counting starts at 0.

```python
text = "Python is amazing!"
print(text.index("is"))
```

27. rindex()
    The method returns the highest index (the one closest to the end) where the substring starts.

```python
text = "apple, banana, apple, cherry"
print(text.rindex("apple"))
```

28. strip()
    It is used to break a single string into a list of smaller strings (substrings).

```python
sentence = "Python is fun"
words = sentence.split()
print(words)
```

29. lstrip()
    It is used to remove characters from the beginning (the left side) of a string.

```python
text = "    Hello World    "
print(text.lstrip() + "|")
text = "---***Important***---"
print(text.lstrip("-*"))
```

30. rsprit()
    This method is the "right-side" version of the standard split() function.

```python
text = "    Hello World    "
print(text.rstrip() + "|")
text = "---***Important***---"
print(text.rstrip("-*"))
```

31. removeprefix()
   It is a string method designed to solve a very specific problem: removing a specific sequence of characters from the beginning of a string without the "character-set".

```python
text = "Python Programming"
print(text.removeprefix("Python"))
```

32. removesuffix()
    It is the surgical tool for removing a specific string from the end of another string.

```python
text = "Python Programming"
print(text.removesuffix("Programming"))
```

33. format()
    the process of inserting variables or values into specific places within a string.

```python
sentence = "My name is {} and I love {}.".format("Shivam", "Python")
print(sentence)
```

34. format_map()
    The format_map() method is a specialized sibling of the format() method. While they do essentially the same thing—mapping values into placeholders—format_map() is designed to work directly with a dictionary (or any mapping object).

```python
data = {'name': 'Shivam', 'role': 'Engineer', 'level': 5}
msg = "User: {name} | Role: {role} | Level: {level}".format_map(data)
print(msg)
```

35. maketrans()
The maketrans() function is a static method used to create a translation table. This table is essentially a "mapping" or a "dictionary" that tells Python which characters should be replaced with which other characters.

```python
table = str.maketrans({"A": "4", "E": "3", "I": "1"})
text = "APPLE"
print(text.translate(table))
```

36. translate()
    It is significantly faster than calling .replace() multiple times because it performs all substitutions in a single pass over the string.

```python
table = str.maketrans("aeiou", "12345")
text = "hello world"
result = text.translate(table)
print(result)
```
